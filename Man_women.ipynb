```python
import zipfile
import os
```

```python
uploaded_zip = "/content/robber dataset.v1-wepon-detection.yolov11.zip"
extracted_path = '/content/dataset'
```

```python
with zipfile.ZipFile(uploaded_zip, 'r') as zip_ref:
    zip_ref.extractall(extracted_path)

print("âœ… Dataset extracted to:", extracted_path)
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")  # small & fast

results = model.train(
    data="/content/robber dataset.v1-wepon-detection.yolov11.zip",
    epochs=50,
    imgsz=640,
    batch=16,
    device=0,
    workers=2,
    patience=10,
    lr0=0.01
)
```
